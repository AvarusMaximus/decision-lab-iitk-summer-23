```python
import pandas as pd
import matplotlib.pyplot as plt
```

## Baseline Flanker Data

```python
#Load the data from the CSV file
baseline_data = pd.read_csv('baseline.csv')
```

```python
#Checking the data
baseline_data.head()
```

```python
print(baseline_data.columns) #Checking the columns in the dataset
```

```python
print(baseline_data['congruent'].value_counts()) #Printing the number of congruent and incongruent trials (1.0: congruent, 0.0: Incongruent)
```

```python
# Separate the data into congruent and incongruent trials
baseline_congruent_trials = baseline_data[baseline_data['congruent'] == 1]
baseline_incongruent_trials = baseline_data[baseline_data['congruent'] == 0]
```

```python
# Check the congruent and incongruent trials
baseline_congruent_trials.head()
```

```python

baseline_incongruent_trials.head()
```

```python
# Check the range of values in the key_resp.rt column
print('Baseline Congruent Trials Mean Reaction Time:', baseline_congruent_trials['key_resp.rt'].mean(), 's')
print('Baseline Congruent Trials Median Reaction Time:', baseline_congruent_trials['key_resp.rt'].median(), 's')
print('Baseline Incongruent Trials Mean Reaction Time:', baseline_incongruent_trials['key_resp.rt'].mean(), 's')
print('Baseline Incongruent Trials Median Reaction Time:', baseline_incongruent_trials['key_resp.rt'].median(), 's')
```

```python
#Creating a box plot of response time for congruent and incongruent trials
plt.boxplot([baseline_congruent_trials['key_resp.rt'].dropna(), baseline_incongruent_trials['key_resp.rt'].dropna()], labels = ['Congruent', 'Incongruent'])
plt.ylabel('Response Time (s)')
plt.title('Response Time for Congruent and Incongruent Trials in Baseline Data')
plt.show()
```

```python
#Separating the Congruent Trials into Correct and Incorrect
baseline_correct_congruent_trials = baseline_congruent_trials[baseline_congruent_trials['key_resp.corr'] == 1]
baseline_incorrect_congruent_trials = baseline_congruent_trials[baseline_congruent_trials['key_resp.corr'] == 0]
```

```python
# Create a box plot of response times for correct and incorrect congruent trials
plt.boxplot([baseline_correct_congruent_trials['key_resp.rt'].dropna(), baseline_incorrect_congruent_trials['key_resp.rt'].dropna()], labels=['Correct', 'Incorrect'])
plt.ylabel('Response Time (s)')
plt.title('Response Times for Correct and Incorrect Congruent Trials for Baseline Data')
plt.show()
```

```python
# Separate the congruent trials into correct and incorrect trials
baseline_correct_incongruent_trials = baseline_incongruent_trials[baseline_incongruent_trials['key_resp.corr'] == 1]
baseline_incorrect_incongruent_trials = baseline_incongruent_trials[baseline_incongruent_trials['key_resp.corr'] == 0]
```

```python
# Create a box plot of response times for correct and incorrect congruent trials
plt.boxplot([baseline_correct_incongruent_trials['key_resp.rt'].dropna(), baseline_incorrect_incongruent_trials['key_resp.rt'].dropna()], labels=['Correct', 'Incorrect'])
plt.ylabel('Response Time (s)')
plt.title('Response Times for Correct and Incorrect Incongruent Trials for Baseline Data')
plt.show()
```

```python
print(baseline_data['key_resp.corr'].value_counts()) #Checking the number of correct and incorrect responses
```

```python
# Create a new column to indicate if the previous trial was an error
baseline_data['prev_error'] = baseline_data['key_resp.corr'].shift(1) == 0
```

```python
# Filter the data to only include correct trials
baseline_correct_trials = baseline_data[baseline_data['key_resp.corr'] == 1]
```

```python
# Create a box plot of response times for correct trials after an error and normal correct trials
plt.boxplot([baseline_correct_trials[baseline_correct_trials['prev_error']]['key_resp.rt'].dropna(), baseline_correct_trials[~baseline_correct_trials['prev_error']]['key_resp.rt'].dropna()], labels=['After Error', 'Normal'])
plt.ylabel('Response Time (s)')
plt.title('Response Times for Correct Trials in Baseline Game')
plt.show()
```

```python
# Calculate the mean response time for correct trials after an error
mean_rt_after_error = baseline_correct_trials[baseline_correct_trials['prev_error']]['key_resp.rt'].mean()

# Calculate the mean response time for normal correct trials (not after an error)
mean_rt_normal = baseline_correct_trials[~baseline_correct_trials['prev_error']]['key_resp.rt'].mean()
```

```python
# Create a bar plot of mean response times for correct trials after an error and normal correct trials
plt.bar(['Correct Response After Error Trials', 'Correct Response After Correct Trials'], [mean_rt_after_error, mean_rt_normal])
plt.ylabel('Mean Response Time (s)')
plt.title('Mean Response Times for Correct Trials In Baseline')
plt.show()
```
