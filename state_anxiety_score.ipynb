```python
import pandas as pd
import numpy as np
```

Loading the data

```python
raw_data = pd.read_csv('Flanker task survey - pilot_July 8, 2023_19.47.csv')
promit_raw_data = pd.read_csv('trait_anx_promit_score.csv')
```

```python
print(promit_raw_data.columns.to_list())
```

```python
state_anx_data = raw_data.iloc[2:, 42:]
state_anx_data
```

```python
print(state_anx_data.columns.to_list().index('State_Anxiety_01.3'))
```

```python
baseline_state_anx = state_anx_data.iloc[:, :21]
baseline_state_anx
```

```python
baseline_state_anx = baseline_state_anx.drop('AttentionCheck_A', axis = 1)
baseline_state_anx
```

```python
pre_stim_state_anx = state_anx_data.iloc[:, 22:43]
pre_stim_state_anx = pre_stim_state_anx.drop('AttentionCheck_A.1', axis = 1)
pre_stim_state_anx
```

```python
post_stim_state_anx = state_anx_data.iloc[:, 43:64]
post_stim_state_anx = post_stim_state_anx.drop('AttentionCheck_A.2', axis = 1)
post_stim_state_anx
```

```python
final_state_anx = state_anx_data.iloc[:, 64:]
final_state_anx = final_state_anx.drop('AttentionCheck_A.3', axis = 1)
final_state_anx
```

```python
state_anx = [baseline_state_anx, pre_stim_state_anx, post_stim_state_anx, final_state_anx]
```

```python
forw_state_index = [2, 3, 5, 6, 8, 11, 12, 13, 16, 17]
rev_state_index = [0, 1, 4, 7, 9, 10, 14, 15, 18, 19]

forw_mapping = dict(zip(["Not at all","Somewhat","Moderately so","Very much so"], [1, 2, 3, 4]))
rev_mapping = dict(zip(["Not at all","Somewhat","Moderately so","Very much so"], [4, 3, 2, 1]))
```

```python
forw_state_anx_lst = []
rev_state_anx_lst = []
forw_state_anx_score = []
rev_state_anx_score = []
state_anx_score = []
for i in range(len(state_anx)):
    forw_state_anx_lst.append(state_anx[i].iloc[:, forw_state_index])
    rev_state_anx_lst.append(state_anx[i].iloc[:, rev_state_index])

    forw_state_anx_lst[i] = forw_state_anx_lst[i].replace(forw_mapping)
    rev_state_anx_lst[i] = rev_state_anx_lst[i].replace(rev_mapping)

    forw_state_anx_score.append(forw_state_anx_lst[i].sum(axis = 1).tolist())
    rev_state_anx_score.append(rev_state_anx_lst[i].sum(axis = 1).tolist())

    state_anx_score.append([forw_state_anx_score[i][x] + rev_state_anx_score[i][x] for x in range(len(forw_state_anx_score[i]))])
    
```

```python
state_anx_score
```
