```python
promit_data = pd.read_csv('trait_anx_promit_score.csv')
promit_data = promit_data.iloc[-2:, 21:42]
promit_data
```

```python
pre_attention = promit_data.iloc[:, :9]
post_attention = promit_data.iloc[:, 10:]
trait_anx = pd.concat([pre_attention, post_attention], axis=1)
trait_df = trait_anx.iloc[:, trait_index_list]
rev_trait_df = trait_anx.iloc[:, rev_trait_index_list]
mapping= dict(zip(["Almost never","Sometimes","Often","Almost Always"],[4,3,2,1]))
trait_df = trait_df.replace(mapping)
trait_df
rev_mapping = dict(zip(["Almost never","Sometimes","Often","Almost Always"],[1,2,3,4]))
rev_trait_df = rev_trait_df.replace(rev_mapping)
rev_trait_df
trait_anx_score = trait_df.sum(axis = 1)
rev_trait_anx_score = rev_trait_df.sum(axis = 1)
trait_anx_score_lst = trait_anx_score.to_list()
rev_trait_anx_score_lst = rev_trait_anx_score.to_list()
total_trait_anx_score = [trait_anx_score_lst[x] + rev_trait_anx_score_lst[x] for x in range(len(trait_anx_score_lst))]
total_trait_anx_score
```

```python
promit_score = np.mean(total_trait_anx_score)
```
